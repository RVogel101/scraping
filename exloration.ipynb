```python
# Remove the copyright notice from extracted text
copyright_text = "The «Word of the Day» is the intellectual property of the Centre for Western Armenian Studies, and any use of it on other platforms or in any other circumstances requires the permission of the Centre."
copyright_text = "Word breakdown օգտ + w+ կար "Օգտ" derives from "օգուտ" (usefulness, benefit) "Կար" is the root of (possible). "Ա" Տ a linking letter. Centre for Western Armenian Studies @ Արեւմտահայագիտական ՈւսմաԿց Կեդրոն"
title_text = "Centre for Western Armenian Studies Արեւմտահայագիտական Ուսմանց Կեդրոն"
add_text = "Centre for Western Armenian Studies | Արեւմտահայագիտական Ուսմանց Կեդրոն WESTERN ARMENIAN"
# Clean the text column
df['text'] = df['text'].str.replace(copyright_text, '', regex=False)
df['text'] = df['text'].str.replace(title_text, '', regex=False)

# Also remove any extra whitespace that might be left
df['text'] = df['text'].str.strip()

print(f"Cleaned text in {len(df)} rows")
print("\nFirst few cleaned entries:")
df.head()
```

```python
# Add category column based on text content
def categorize_text(text):
    text_lower = str(text).lower()
    
    if add_text.lower() in text_lower:
        return 'Add'
    elif "Vocabulary Enhancement Course".lower() in text_lower:
        return 'Add'
    elif 'etymology' in text_lower:
        return 'Etymology'
    elif 'word breakdown' in text_lower:
        return 'Word Breakdown'
    elif 'phrasal breakdown' in text_lower:
        return 'Phrasal Breakdown'
    elif 'declension' in text_lower and 'example' in text_lower:
        return 'Declension-Example'
    elif 'example' in text_lower:
        return 'Example'
    elif 'declension' in text_lower:
        return 'Declension'
    elif 'conjunction' in text_lower:
        return 'Conjunction'
    elif 'conjugation' in text_lower:
        return 'Conjugation'
    elif 'present tense & immediate present tense' in text_lower:
        return "Conjunction"
    else:
        return 'Other'

# Apply categorization
df['category'] = df['text'].apply(categorize_text)

# Show category distribution
print("Category distribution:")
print(df['category'].value_counts())
print(f"\nTotal entries: {len(df)}")

# Show a few examples of each category
print("\nSample entries by category:")
for category in df['category'].unique():
    print(f"\n--- {category} ---")
    sample = df[df['category'] == category].head(2)
    for idx, row in sample.iterrows():
        print(f"Text preview: {str(row['text'])[:100]}...")
```

```python
# Split Declension-Example rows using pandas operations
import re

# Create a function to extract text after "example"
def extract_after_example(text):
    """Extract text that comes after 'example' (case insensitive)"""
    if pd.isna(text) or text == '':
        return ''
    
    text_str = str(text)
    # Find "example" case insensitive
    match = re.search(r'\bexample\b', text_str, re.IGNORECASE)
    
    if match:
        return text_str[match.end():].strip()
    return ''

def extract_before_example(text):
    """Extract text that comes before 'example' (case insensitive)"""
    if pd.isna(text) or text == '':
        return text
    
    text_str = str(text)
    # Find "example" case insensitive
    match = re.search(r'\bexample\b', text_str, re.IGNORECASE)
    
    if match:
        return text_str[:match.start()].strip()
    return text_str

# Filter rows with "Declension-Example" category
declension_example_mask = df['category'] == 'Declension-Example'
print(f"Found {declension_example_mask.sum()} rows with 'Declension-Example' category")

if declension_example_mask.sum() > 0:
    # Create new rows for the "Example" part
    example_rows = df[declension_example_mask].copy()
    example_rows['category'] = 'Example'
    example_rows['text'] = example_rows['text'].apply(extract_after_example)
    
    # Update original rows to contain only the declension part
    df.loc[declension_example_mask, 'category'] = 'Declension'
    df.loc[declension_example_mask, 'text'] = df.loc[declension_example_mask, 'text'].apply(extract_before_example)
    
    # Concatenate the new example rows to the dataframe
    df = pd.concat([df, example_rows], ignore_index=True)
    
    print(f"✓ Split {declension_example_mask.sum()} rows")
    print(f"✓ Updated original rows to 'Declension' category")
    print(f"✓ Created {len(example_rows)} new 'Example' rows")
    print(f"✓ New total rows: {len(df)}")
    
    # Show sample results
    print("\nSample of updated rows:")
    declension_rows = df[df['category'] == 'Declension'].head(2)
    example_rows_sample = df[df['category'] == 'Example'].head(2)
    
    print("\nDeclension rows:")
    for idx, row in declension_rows.iterrows():
        print(f"Text: {str(row['text'])[:100]}...")
    
    print("\nExample rows:")
    for idx, row in example_rows_sample.iterrows():
        print(f"Text: {str(row['text'])[:100]}...")
    
else:
    print("No 'Declension-Example' rows found to split")

# Show updated category distribution
print(f"\nUpdated category distribution:")
print(df['category'].value_counts())
```
